```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('countries.csv')

df
```

```python
df.columns
```

```python
df2=df[['Country', 'Region', 'population', 'Area (sq. mi.)','Literacy (%)']]
df2.head()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
df2.isna().sum()
```

```python
df2.head()
```

```python
df2.rename(columns={'Area (sq. mi.)':'Area', 'Literacy (%)':'Literacy','population':'Population'}, inplace=True)
df2
```

```python
#remove commans from literacy
df2['Literacy'] = df2['Literacy'].str.split(',').str[0]
df2['Literacy'] = df2['Literacy'].astype(int)

df2.head()
```

```python
x=df2['Region']
y=df2['Literacy']

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=x, y=y)
plt.xticks(rotation=90)
```

```python
category_regions=df2.groupby('Region').agg({'Literacy':'mean'}).reset_index()

category_regions

avg_pop=df2.groupby('Region').agg({'Population':'mean'}).reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Literacy', data=category_regions)
plt.xticks(rotation=90)
plt.title('Average Literacy Rate by Region')
```

```python
category_regions=df2.groupby('Region').agg({'Literacy':'mean'}).reset_index()

category_regions

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Region', y='Literacy', data=category_regions)
plt.xticks(rotation=90)
plt.title('Average Literacy Rate by Region')
```

```python
category_regions
```

```python
avg_pop=df2.groupby('Region').agg({'Population':'mean'}).reset_index()
avg_pop
```

# Matplolib

```python
#create multi line plot
plt.figure(figsize=(10, 6))

figure, ax = plt.subplots()
figure.set_size_inches(10, 10) #new way of doign size
ax.plot(category_regions['Region'], category_regions['Literacy'])
ax.plot(avg_pop['Region'], avg_pop['Population'], label='Population')

plt.show()
```

```python
#create multi line plot
plt.figure(figsize=(10, 6))

figure, [ax1,ax2] = plt.subplots(nrows=1, ncols=2)

ax1.plot(category_regions['Region'], category_regions['Literacy'])
ax2.plot(avg_pop['Region'], avg_pop['Population'], label='Population')

plt.show()
```

```python
#create multi line plot
plt.figure(figsize=(10, 10))
figure, ax1 = plt.subplots()

ax1.set_xlabel('Regions')
ax1.set_ylabel('Literacy Rate')


plt.title('Country stats')
ax2= ax1.twinx()
ax2.set_ylabel('Population')

ax1.set_xticklabels(category_regions['Region'], rotation=90)
ax1.plot(category_regions['Region'], category_regions['Literacy'],color='blue', label='Literacy Rate')
ax2.plot(avg_pop['Region'], avg_pop['Population'], label='Population', color='orange')

figure.legend(loc='upper right')

plt.show()
```

```python
# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

plt.figure(figsize=(10, 6))
# Draw a combo histogram and scatterplot with density contours
figure, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
plt.title('Scatterplot with density contours')

plt.savefig('new.png')
plt.show()
```
