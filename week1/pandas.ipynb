```python
df5=df3.query('Population<40000000 and Population>1000000')

df5.plot(kind='box', figsize=(10, 5), title='Population Boxplot')
```

```python
df5['Region']=df5['Region'].str.strip()

df5.sort_values(by='Population', ascending=False, inplace=True)
df5
```

```python
df5['Region'].unique()
```

```python
#calculate the mean population by region

region_data=df5.groupby('Region')
region_data

results=region_data['Population'].mean()

results
```

```python
results.plot(kind='pie', figsize=(10, 5), title='Mean Population by Region')
```
